```python
import pandas as pd

df = pd.read_csv("../data/Superstore.csv", encoding="latin-1")
df.head(10)
```

```python
df.columns
```

```python
# Group by Product Name and sum Sales
top_products = df.groupby('Product Name')['Sales'].sum()

# Sort in descending order
top_products = top_products.sort_values(ascending=False)

# Show top 10 products
top_products.head(10)
```

```python
top_products_df = top_products.reset_index()
top_products_df.columns = ['Product Name', 'Total Sales']
top_products_df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_10 = top_products_df.head(10)
top_10
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x='Total Sales', 
    y='Product Name', 
    data=top_10,
    hue='Product Name',  # assign y variable to hue
    dodge=False,          # keep bars as single
    palette='viridis',
    legend=False          # hide legend since it's obvious
)

plt.title('Top 10 Products by Total Sales', fontsize=16)
plt.xlabel('Total Sales ($)', fontsize=12)
plt.ylabel('Product Name', fontsize=12)
plt.show()
```

```python
# Group by Category and sum Sales & Profit
category_df = df.groupby('Category').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

category_df
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x='Profit', 
    y='Category', 
    data=category_df,
    palette='coolwarm'
)

plt.title('Profit by Category', fontsize=16)
plt.xlabel('Profit ($)', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

```python
# Add a column to mark loss-making categories
category_df['Status'] = category_df['Profit'].apply(lambda x: 'Loss' if x < 0 else 'Profit')

category_df
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x='Profit',
    y='Category',
    data=category_df,
    palette='coolwarm'
)

plt.title('Profit by Category', fontsize=16)
plt.xlabel('Profit ($)', fontsize=12)
plt.ylabel('Category', fontsize=12)

plt.savefig('../output/category_profit.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd

# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create 'Month-Year' column
df['Month-Year'] = df['Order Date'].dt.to_period('M')

# Aggregate sales by Month-Year
monthly_sales = df.groupby('Month-Year')['Sales'].sum().reset_index()

# Convert Month-Year back to datetime for plotting
monthly_sales['Month-Year'] = monthly_sales['Month-Year'].dt.to_timestamp()

monthly_sales.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(x='Month-Year', y='Sales', data=monthly_sales, marker='o')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.savefig('../output/Monthly_Sales_trends.png', bbox_inches='tight')
plt.show()
```

```python
# 3-month moving average
monthly_sales['Sales_MA_3'] = monthly_sales['Sales'].rolling(window=3).mean()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales['Month-Year'], monthly_sales['Sales'], marker='o', label='Actual Sales')
plt.plot(monthly_sales['Month-Year'], monthly_sales['Sales_MA_3'], marker='o', label='3-Month MA Forecast')
plt.title('Sales Forecast using Moving Average')
plt.xticks(rotation=45)
plt.legend()
plt.savefig('../output/Sales_Forecast_Using_Moving_Average.png', bbox_inches='tight')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Convert Month-Year to numerical values
monthly_sales['Month_Num'] = np.arange(len(monthly_sales))

# Fit Linear Regression
X = monthly_sales[['Month_Num']]
y = monthly_sales['Sales']

model = LinearRegression()
model.fit(X, y)

# Predict next month
next_month_num = np.array([[len(monthly_sales)]])
next_month_sales = model.predict(next_month_num)[0]

print(f"Forecasted Sales for Next Month: {next_month_sales:.2f}")

# Plot
plt.figure(figsize=(12,6))
plt.plot(monthly_sales['Month-Year'], y, marker='o', label='Actual Sales')
plt.plot(monthly_sales['Month-Year'], model.predict(X), marker='o', label='Linear Regression Fit')
plt.title('Sales Forecast using Linear Regression')
plt.xticks(rotation=45)
plt.legend()
plt.savefig('../output/Sales_Forecast_Using_Linear_Regression.png', bbox_inches='tight')
plt.show()
```
